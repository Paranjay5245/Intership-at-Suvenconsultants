#### (1) Import require libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings 
warnings.filterwarnings('ignore')
```

#### (2) Read Dataset

```python
df=pd.read_csv("D:\\study\\weatherHistory.csv")
```

#### (4) Before moving on visulization, we have to convert formatted date to datetype object.

```python
df["Formatted Date"]=pd.to_datetime(df["Formatted Date"],utc=True)
```

```python
df.dtypes
```

#### After converting object to datetime

```python
df.head()
```

```python
df["Formatted Date"]
```

#### (5) Now, here data is in hourly period of time. so, for better analysis we need to resample our data in monthly.

```python
#set formatted Date as an index
df=df.set_index("Formatted Date")
```

```python
df.head(4)
```

```python
data_columns = ['Apparent Temperature (C)', 'Humidity']
df_ms = df[data_columns].resample('MS').mean()
df_ms.head()
```

# ![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

# (6) Plotting of data

```python
plt.figure(figsize=(20,10))
plt.title("Relation of Apparent Temperature (C) and Humidity",fontsize=20)
sns.lineplot(x="Apparent Temperature (C)",y="Humidity",data=df_ms,color="k")
```

# ![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

# Plot of Apparent Temperature (C) and Humidity with year

```python
plt.figure(figsize=(20,10))
plt.title("Plot of Apparent Temperature (C) and Humidity with Year",fontsize=20)
plt.plot(df_ms["Apparent Temperature (C)"],"violet",label="Apparent Temperature (C)")
plt.plot(df_ms["Humidity"],"purple",label="Humidity")
plt.xlabel("Year",fontsize=18)
plt.legend(fontsize='x-large')
```

# ![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

### we need to find whether the for the month of a month say April starting from 2006 to 2016 and the for the same period have increased or not.

```python
April=df_ms[df_ms.index.month==4]
```

```python
April
```

```python
plt.figure(figsize=(16,7))
plt.title("Plot of Apparent Temperature (C) and Humidity with year for April month",fontsize=16,color="k")
plt.xlabel("April Month",fontsize=18)
plt.plot(April["Apparent Temperature (C)"],"violet",marker="o",label="Apparent Temperature (C)")
plt.plot(April["Humidity"],color='purple',marker="o",label="Apparent Temperature (C)")
plt.legend(fontsize="x-large")
```

#### According to visulization of April month graph for year 2006 to 2016, the average humidity for all period is same. it not increasing nor descreasing. But there is vast change in temperature for 2008, 2009 and 2015 year majorly.

### This monthly analysis has to be done for all 12 months over the 10 year period. So basically after resampling data from hourly to monthly, then comparing the same month over the 10 year period.

```python
Jan=df_ms[df_ms.index.month==1]
Feb=df_ms[df_ms.index.month==2]
March=df_ms[df_ms.index.month==3]
April=df_ms[df_ms.index.month==4]
May=df_ms[df_ms.index.month==5]
June=df_ms[df_ms.index.month==6]
July=df_ms[df_ms.index.month==7]
August=df_ms[df_ms.index.month==8]
Sept=df_ms[df_ms.index.month==9]
Octo=df_ms[df_ms.index.month==10]
Nov=df_ms[df_ms.index.month==11]
Dec=df_ms[df_ms.index.month==12]
```

```python
plt.figure(figsize=(20,10))
plt.title("Plot of Humidity for all year based on month",fontsize=16)
plt.plot(Jan['Humidity'],label='January',color='Purple',marker='o')
plt.plot(Feb['Humidity'],label='Febuary',color='red',marker='o')
plt.plot(March['Humidity'],label='March',color='black',marker='o')
plt.plot(April['Humidity'],label='April',color='yellow',marker='o')
plt.plot(May['Humidity'],label='May',color='blue',marker='o')
plt.plot(June['Humidity'],label='June',color='orange',marker='o')
plt.plot(July['Humidity'],label='July',color='pink',marker='o')
plt.plot(August['Humidity'],label='August',color='green',marker='o')
plt.plot(Sept['Humidity'],label='September',color='brown',marker='o')
plt.plot(Octo['Humidity'],label='Octomber',color='violet',marker='o')
plt.plot(Nov['Humidity'],label='November',color='black',marker='o')
plt.plot(Dec['Humidity'],label='December',color='silver',marker='o')
plt.legend()
```
